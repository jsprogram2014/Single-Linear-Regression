```python
# Linear Regression Model Creation

import numpy as np
import matplotlib.pyplot as matplt
matplt.style.use('./deeplearning.mplstyle')

x_var=np.array([1,2])
y_var=np.array([300,500])

# print(f"The length of the training set is {x_var.shape[0]}")
# print(f"The dimension of the training set is {x_var.shape}")

#function to predict output
def compute_linear_output(x,w,b):
    m=x.shape[0]
    linear_output=np.zeros(m) ## creating a 1 dimensional array with length m
    
    for i in range(m):
        linear_output[i]=w*x[i]+b

    return linear_output

# print(f"Linear Outputs are {compute_linearoutput(x_var,10,50)}")
lnr_opt=compute_linear_output(x_var,10,50)
        
#code to create the scatter plot
# matplt.scatter(x_var,y_var,marker='x',c='r',label='Actual Value')
# matplt.plot(x_var,lnr_opt,c='b',label='Our Prediction')
# matplt.title('Linear Regression Model for House Price Prediction')
# matplt.xlabel('House(1000 sqft)')
# matplt.ylabel('Price(1000 dollars)')
# matplt.legend()
# matplt.show()

def compute_cost(x,y,w,b):
    m=x_var.shape[0]
    cost=0
    for i in range(m):
        fwb=w*x[i]+b
        cost=cost + (fwb-y[i])**2
    total_cost=(1/(2*m))*cost

    return total_cost

# print(f'Average Error is {compute_cost(x_var,y_var,10,50)}')

# code to find dj_dw & dj_db 
def compute_gradient(x,y,w,b):
    m=x.shape[0]
    dj_dw=0
    dj_db=0
    
    for i in range(m):
        dj_dw=dj_dw + ((w*x[i]+b)-y[i])*x[i]
        dj_db=dj_db + (w*x[i]+b)-y[i] 

    dj_dw=(1/m)*dj_dw
    dj_db=(1/m)*dj_db

    return dj_dw,dj_db

def minimize_cost_function():
    w=10
    b=10
    a=0.0001
    i=0
    m=x_var.shape[0]
    while(i<=10000):
        dj_dw,dj_db=compute_gradient(x_var,y_var,w,b)
        tempw=w-a*dj_dw
        tempb=b-a*dj_db
        cost=compute_cost(x_var,y_var,tempw,tempb)
        if(tempw==w):
            break
        w=tempw
        b=tempb
        i+=1
    lnropt2=np.zeros(m)
    lnropt2=compute_linear_output(x_var,w,b)

    #code to create the scatter plot
    matplt.scatter(x_var,y_var,marker='x',c='r',label='Actual Value')
    matplt.plot(x_var,lnropt2,c='b',label='Our Prediction')
    matplt.title('Linear Regression Model for House Price Prediction')
    matplt.xlabel('House(1000 sqft)')
    matplt.ylabel('Price(1000 dollars)')
    matplt.legend()
    matplt.show()
    print(f'The 1400 sq ft house price costs {w*1.4+b}')

minimize_cost_function()

        
    

    
```
